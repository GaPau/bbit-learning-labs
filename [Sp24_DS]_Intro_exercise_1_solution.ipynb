## Problem 1

```python
def rectangle_area(x0,y0,x1,y1):
    #print(x0,y0,x1,y1)
    return (x1-x0)*(y1-y0)

def overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1):
    overlap_x0 = max(r0x0,r1x0)
    overlap_x1 = min(r0x1,r1x1)
    overlap_y0 = max(r0y0,r1y0)
    overlap_y1 = min(r0y1,r1y1)
    if overlap_x1 < overlap_x0:
        return 0
    if overlap_y1 < overlap_y0:
        return 0
    return rectangle_area(overlap_x0,overlap_y0,overlap_x1,overlap_y1)

print('----------------------')
(r0x0,r0y0) = (-4,2)
(r0x1,r0y1) = (3,5)

(r1x0,r1y0) = (1,-2)
(r1x1,r1y1) = (7,4)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)

print('----------------------')
(r0x0,r0y0) = (1,-2)
(r0x1,r0y1) = (7,4)

(r1x0,r1y0) = (-4,2)
(r1x1,r1y1) = (7,4)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)


print('----------------------')
(r0x0,r0y0) = (1,-2)
(r0x1,r0y1) = (7,4)

(r1x0,r1y0) = (1,-2)
(r1x1,r1y1) = (7,4)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)




print('----------------------')
(r0x0,r0y0) = (1,-2)
(r0x1,r0y1) = (7,6)

(r1x0,r1y0) = (2,2)
(r1x1,r1y1) = (5,5)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)


print('----------------------')
(r0x0,r0y0) = (4,-2)
(r0x1,r0y1) = (7,4)

(r1x0,r1y0) = (-4,2)
(r1x1,r1y1) = (3,5)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)



print('----------------------')
(r0x0,r0y0) = (1,-2)
(r0x1,r0y1) = (7,4)

(r1x0,r1y0) = (-4,-3)
(r1x1,r1y1) = (8,5)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)


print('----------------------')
(r0x0,r0y0) = (5,6)
(r0x1,r0y1) = (7,16)

(r1x0,r1y0) = (8,17)
(r1x1,r1y1) = (18,25)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)



print('----------------------')

(r0x0,r0y0) = (8,17)
(r0x1,r0y1) = (18,25)

(r1x0,r1y0) = (5,6)
(r1x1,r1y1) = (7,16)

a0 = rectangle_area(r0x0,r0y0,r0x1,r0y1)
print('Area rectangle 0',a0)

a1 = rectangle_area(r1x0,r1y0,r1x1,r1y1)
print('Area rectangle 1',a1)

a2 = overlap(r0x0,r0y0,r0x1,r0y1,r1x0,r1y0,r1x1,r1y1)
print('Area overlap',a2)

print('Total area',a0+a1-a2)
```

## Problem 2

```python
def hour_angle(h,m):
    h = h%12 + m/60
    return h*360/12

def minute_angle(h,m):
    return m*360/60

def minimum_angle(h,m):
    ha = hour_angle(h,m)
    ma = minute_angle(h,m)
    d = abs(ha-ma)
    return min(d,360-d)

for h in range(12):
    if h==0:h=12
    for m in range(0,60,5):
        print(f'{h = }, {m = }, {minimum_angle(h,m) = :.2f}')
```
